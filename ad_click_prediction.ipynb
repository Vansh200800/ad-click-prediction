### Now we have done we exploratory data analysis we will now create clustering model to indentify the groups of customers based on Age and Area Income

```python
#Selecting subset of columns to perform clustering
df_clustering=df[['Age','Area Income']]
```

```python
df_clustering
```

```python
# converting dataframe into array
X =df_clustering.values
```

Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

The optimal number of cluster are 4

Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('AGE')
plt.ylabel('INCOME')
plt.legend()
plt.show()
```

# Preparing Features for classification model

### Encoding Categorical features 

```python
# We will now Encode Categorical features  
from sklearn import preprocessing 
label = preprocessing.LabelEncoder() 
df['Ad Topic Line']= label.fit_transform(df['Ad Topic Line'])
df['City']= label.fit_transform(df['City'])
df['Country']= label.fit_transform(df['Country'])
```

### Feature Selection

```python
from sklearn.ensemble import ExtraTreesClassifier
model=ExtraTreesClassifier()
X=df.drop(['Clicked on Ad'],axis=1)
y=df['Clicked on Ad']
```

```python
model.fit(X,y)
```

```python
feat_importance=pd.Series(model.feature_importances_,index=X.columns)
```

```python
plt.figure(figsize=(15,6))
feat_importance.sort_values(ascending=True).plot(kind='barh');
```

We applied feature importances in order to check which features are more important in order to predict dependent variable.
We will use four most important features in order to build our future machine learning model

```python
df=df[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Clicked on Ad']]
df.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
# We will use 80 % of data for traning and remaning 20 for testing purpose
X = df.drop('Clicked on Ad',axis=1)
y = df['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Logistics Regression

```python
from sklearn.linear_model import LogisticRegression
```

### K-fold Cross validation

Using K-fold cross validation to compute accuracy using LogisticRegression

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = LogisticRegression(), X = X_train, y = y_train, cv = 5)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
```

### Grid Search Cross validation 

```python
from sklearn.model_selection import GridSearchCV
#Using Grid Search CV to get best parameters
classfier = GridSearchCV(LogisticRegression(), {                                          
    'C':[.5, 1, 5, 10,20],
    'solver' : ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
    
}, cv=3, return_train_score=False)
classfier.fit(X_train,y_train)
```

```python
print("Best Accuracy: {:.2f} %".format(classfier.best_score_*100))
print("Best Parameters:", classfier.best_params_)
```

We have obtained a very robust and accurate model that can provide accurate predictions upto 97%

```python
classifer=LogisticRegression(C=0.5,solver='newton-cg')
classifer.fit(X_train,y_train)
```

```python
y_pred=classifer.predict(X_test)
```

### Printing Confusion matrix 

```python
from sklearn.metrics import confusion_matrix
```

```python
Confusion_matrix = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(Confusion_matrix,annot=True,fmt='.0f');
```

### Classifier report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

# Artifical Neural Network

```python
import tensorflow as tf
```

Building the ANN

```python
#Initializing the ANN
ann = tf.keras.models.Sequential()
#Adding the input layer and the first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
#Adding the output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Training the ANN

```python
#Compiling the ANN
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
#Training the ANN on the Training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

Predicting the Test set

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
#Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True,fmt='.0f');
```

```python
print("Accuracy of Neural Network : ",accuracy_score(y_test, y_pred))
```
